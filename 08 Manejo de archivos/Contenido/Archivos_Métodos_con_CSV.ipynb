# MÃ©todos de escritura
## write()
Este mÃ©todo es como ya venÃ­amos viendo.
* Escribe una sola cadena de texto en el archivo.
* No agrega el salto de lÃ­nea (\n) automÃ¡ticamente.
* Se usa para registrar datos uno por uno.

```python
print("Escribiendo con write()")

with open("productos.csv", "w") as archivo:
    archivo.write("Nombre,Precio,Cantidad\n")
    archivo.write("Cartuchera,750.0,10\n")
    archivo.write("Regla,100.0,25\n")
```

## writelines()
Escribe una lista de lÃ­neas de texto. No agrega saltos de lÃ­nea automÃ¡ticamente, por lo que deben estar incluidos.
* Escribe una lista de cadenas de texto de una sola vez.
* Tampoco agrega saltos de lÃ­nea automÃ¡ticamente, asÃ­ que hay que incluirlos manualmente.
* Es Ãºtil para escribir varias lÃ­neas juntas de forma rÃ¡pida.

```python
print("Escribiendo con writelines():")

lineas = [
    "Nombre,Precio,Cantidad\n",
    "CompÃ¡s,180.0,12\n",
    "Tijera,250.0,20\n"
]

with open("productos.csv", "w") as archivo:
    archivo.writelines(lineas)
```

1.  DefiniciÃ³n de la lista lineas
* Creamos una lista de cadenas de texto, donde cada elemento representa una lÃ­nea del archivo.
* El primer string es el encabezado (los nombres de las columnas).
* Los demÃ¡s strings son registros, con valores separados por comas.
* \n es el carÃ¡cter que indica un salto de lÃ­nea, para que cada string se escriba en una lÃ­nea separada.
2. Apertura del archivo con with open(...) en modo escritura
* Abre (o crea si no existe) el archivo mas_productos.csv en modo de escritura ("w").

3.  Escritura con writelines()
* Escribe todas las lÃ­neas de la lista lineas dentro del archivo.
* No agrega saltos de lÃ­nea automÃ¡ticamente, por eso ya estaban incluidos (\n) en cada string de la lista.
* El resultado serÃ¡ un archivo con 3 lÃ­neas: el encabezado y dos productos.

#MÃ©todos de Lectura##
###readline()
Lee una sola lÃ­nea del archivo por cada llamada.
* Lee una sola lÃ­nea del archivo por vez.
* Cada vez que lo llamamos, avanza una lÃ­nea mÃ¡s.
* Ideal cuando queremos procesar el archivo lÃ­nea por lÃ­nea de forma controlada.

```python
with open("productos.csv", "r") as archivo:
    encabezado = archivo.readline().strip()
    print("Encabezado:", encabezado)

    linea1 = archivo.readline().strip()
    print("Primera lÃ­nea:", linea1)

    linea2 = archivo.readline().strip()
    print("Segunda lÃ­nea:", linea2)
```

1. with open("productos.csv", "r") as archivo:
* Abrimos el archivo productos.csv en modo lectura ("r").
* Gracias a with, no es necesario cerrar el archivo manualmente; Python lo hace al salir del bloque.
* La variable archivo representa la conexiÃ³n con el archivo y nos permite leer su contenido.

2. archivo.readline()
* Este mÃ©todo lee una sola lÃ­nea del archivo por cada vez que se lo llama.
* Se mueve hacia la siguiente lÃ­nea automÃ¡ticamente en cada llamada.
* Por eso en este cÃ³digo hay tres llamadas a readline(), una para el encabezado y dos para los registros.

3. .strip()
* Se utiliza para eliminar el salto de lÃ­nea final (\n) y posibles espacios en blanco al principio o final de cada lÃ­nea.
* Es buena prÃ¡ctica para trabajar con los datos limpios.

4. encabezado = archivo.readline().strip()
* Lee la primera lÃ­nea del archivo, que normalmente contiene los nombres de las columnas (por ejemplo, "Nombre,Precio,Cantidad").
* Se guarda en la variable encabezado y se imprime.

5. linea1 = archivo.readline().strip()
* Lee la segunda lÃ­nea, es decir, el primer producto del archivo.
* Por ejemplo: "Lapicera,120.5,30"

6. linea2 = archivo.readline().strip()
* Lee la tercera lÃ­nea, o sea, el segundo producto.

### realines()
Lee todas las lÃ­neas del archivo y las guarda en una lista de strings.
* Cada lÃ­nea se convierte en un elemento de la lista.
* Ãtil cuando queremos trabajar con todo el contenido al mismo tiempo.

```python
print("Leyendo con readlines():")

with open("productos.csv", "r") as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        partes = linea.strip().split(",")
        print(partes)
```

1. with open("productos.csv", "r") as archivo:
* Abre el archivo productos.csv en modo lectura ("r").
* Se usa with para asegurar que el archivo se cierre automÃ¡ticamente al finalizar el bloque.
* archivo es la variable que representa la conexiÃ³n con el archivo.

2. lineas = archivo.readlines()
* El mÃ©todo .readlines() lee todas las lÃ­neas del archivo y devuelve una lista, donde cada elemento es una lÃ­nea como cadena de texto.

3. for linea in lineas:
* Recorre cada lÃ­nea de la lista lineas.

4. linea.strip()
* Elimina los caracteres innecesarios al principio o final de la lÃ­nea, como el salto de lÃ­nea \n.
* Ejemplo: "Lapicera,120.5,30\n" se convierte en "Lapicera,120.5,30".

5. .split(",")
* Divide la lÃ­nea en partes, separÃ¡ndola por las comas.
* Devuelve una lista con los elementos de la lÃ­nea.
* Ejemplo: "Lapicera,120.5,30" se transforma en ["Lapicera", "120.5", "30"].

6. print(partes)
* Muestra en pantalla cada lÃ­nea ya procesada en forma de lista.

**Este cÃ³digo es ideal para leer todos los registros de un archivo CSV simple y convertir cada lÃ­nea en una lista de datos Ãºtiles que se pueden manipular fÃ¡cilmente dentro del programa.**
