Un archivo CSV (Comma-Separated Values, o "valores separados por comas") es un tipo de archivo de texto plano que almacena datos en forma tabular, similar a una hoja de clculo.

* Cada lnea del archivo representa una fila de la tabla.

* Cada valor, separado por comas, representa una columna.

Motivos de su uso:
* Son simples de leer y escribir.

* Son compatibles con casi todos los programas de hojas de clculo (como Excel o Google Sheets).

* Son una forma muy comn de intercambiar datos entre sistemas.

# Escritura de archivos CSV

````python
with open("productos.csv", "w") as archivo:
    archivo.write("Lapicera,120.5,30\n")
    archivo.write("Cuaderno,300.0,15\n")
    archivo.write("Goma,80.0,50\n")
````

# Lectura de un archivo CSV


````python
print("Leyendo archivo CSV y separando datos")
with open("productos.csv", "r") as archivo:
    for linea in archivo:
        linea = linea.strip()
        partes = linea.split(",")
        print(partes)
````

 1. with open("productos.csv", "r") as archivo:
    * Abre el archivo productos.csv en modo lectura ("r").
    * Se usa with para que el archivo se cierre automticamente cuando termine el bloque.
    * La variable archivo es el pseudnimo que vamos a usar para trabajar con el archivo dentro del bloque.

2. for linea in archivo:
    * Recorre lnea por lnea el contenido del archivo.
    * Cada linea es una cadena de texto que representa una fila del archivo.

3. linea = linea.strip() :
    * Elimina espacios vacos al principio o final de la lnea, especialmente el salto de lnea (\n) al final de cada lnea del archivo.

4. partes = linea.split(",") :
   * Divide la lnea en una lista de elementos, separados por comas (,), que es el formato estndar de los archivos CSV.
   * Por ejemplo: "Lapicera,120.5,30" se convierte en ['Lapicera', '120.5''30'].

5. print(partes) :
    * Muestra en pantalla la lista resultante, con los valores ya separados y listos para procesar.

# Funcin strip()
Se utiliza para eliminar los caracteres vacos al principio y al final de una cadena de texto.
Caracteres que elimina:
* Espacios
* Tabulaciones (\t)
* Saltos de lnea (\n)

Por eso es muy til cuando leemos lneas desde un archivo, ya que suelen venir con un salto de lnea al final.

```python
texto = "  Hola mundo\n" #  La frase posee espacios al principio y salto de lnea al final
print(texto)             #  Con espacios y salto
print(texto.strip())     #  'Hola mundo'


# Funcin split()

 Se utiliza para dividir una cadena de texto en partes, generando una lista. Por defecto, divide usando espacios, pero tambin puede dividir por comas, puntos, guiones, etc.


```python
frase = "Hola cmo ests"  # Toda la frase est en una nica cadena de texto (string)
palabras = frase.split()   # split() separa la cadena por espacios y devuelve una lista de palabras
print(palabras)            # ['Hola', 'cmo', 'ests']
