# Escritura inicial de un archivo

```python
print("1. Escribiendo archivo")
archivo_productos = open("productos.txt", "w")
archivo_productos.write("Lapicera\n")
archivo_productos.write("Cuaderno\n")
archivo_productos.close()
```



1. open("productos.txt", "w") es la funci√≥n que permite abrir un archivo.
En este caso, se guarda su resultado en la variable archivo_productos, que representar√° el archivo mientras lo usamos.
Recibe dos par√°metros:

    * "productos.txt": es el nombre o ruta del archivo que queremos abrir o crear.

    * "w": es el modo de apertura. En este caso 'w' significa write (escribir).
    Si el archivo ya existe, se sobrescribe. Si no existe, se crea.

2. archivo_productos.write("Lapicera\n") es el m√©todo que se usa para escribir texto dentro del archivo.
Se llama desde la variable que representa el archivo (archivo_productos) y recibe como par√°metro el texto a guardar.

    En este caso, cada l√≠nea termina con \n que representa un salto de l√≠nea.

3. archivo_productos.close() se encarga de cerrar el archivo cuando terminamos de usarlo.
Esto es muy importante para asegurarnos de que todo lo escrito se guarde correctamente y para liberar recursos del sistema.

# Lectura de un archivo

```python
print("2. Leyendo archivo y mostrando contenido")
archivo_productos = open("productos.txt", "r")
contenido = archivo_productos.read()
print(contenido)
archivo_productos.close()
```

1. open("productos.txt", "r") es la funci√≥n que permite abrir un archivo existente para lectura.

    "productos.txt" es el nombre del archivo que queremos leer.

    "r" es el modo de apertura (read, es decir, solo lectura).
    El archivo debe existir, si no, se generar√° un error.

  El resultado se guarda en la variable archivo, que usaremos para acceder al contenido.

2. archivo.read() es el m√©todo que lee todo el contenido del archivo de una sola vez y lo devuelve como una cadena de texto.
Este contenido se guarda en la variable contenido.

3. print(contenido) simplemente muestra en pantalla lo que se ley√≥ del archivo.
Si el archivo ten√≠a varias l√≠neas, se ver√°n con los saltos de l√≠nea originales.

4. archivo.close() cierra el archivo luego de haber terminado la lectura.
Siempre es buena pr√°ctica cerrar los archivos manualmente si no usamos with, para evitar errores o consumo innecesario de recursos.

# Insertar registros al final


```python
print("3. Agregando contenido al final del archivo")
archivo_productos = open("productos.txt", "a")
archivo_productos.write("Goma de borrar\n")
archivo_productos.close()
```

1. open("productos.txt", "a") abre el archivo en modo de agregado (append).

    "productos.txt" es el archivo al que queremos a√±adir contenido.

    "a" indica que queremos agregar al final del archivo sin borrar lo que ya tiene.
    Si el archivo no existe, se crea autom√°ticamente.

  La funci√≥n devuelve un objeto que se guarda en la variable archivo_productos, desde la cual se puede escribir.

2. archivo_productos.write("Goma de borrar\n") escribe una nueva l√≠nea al final del archivo.

    El texto se agrega despu√©s del contenido ya existente.

    Se usa \n para agregar un salto de l√≠nea, asegurando que la l√≠nea no se pegue con la anterior.

3. archivo_productos.close() cierra el archivo una vez finalizada la operaci√≥n.
Esto es importante para que los datos se guarden correctamente y se liberen los recursos del sistema.

# Lectura segura con with y modo 'r'
Python tambi√©n nos ofrece una forma m√°s segura de manipular archivos mediante la sentencia with. Esta estructura evita tener que cerrar manualmente la conexi√≥n cada vez que abrimos un archivo, lo cual es √∫til en caso de olvidos o errores. Al utilizar with, el cierre del archivo se realiza de forma autom√°tica al finalizar el bloque de c√≥digo.

```python
print("Leyendo archivo con with")
with open("productos.txt", "r") as archivo:
    contenido = archivo.read()
    print(contenido)
```

1. El bloque with open(...) as archivo: abre el archivo y se asegura de cerrarlo autom√°ticamente al finalizar el bloque.

2. "r" es el modo de lectura, usado para leer el contenido existente.

3. La variable archivo se puede usar solo dentro del bloque.

4. Es una forma m√°s segura y recomendada para trabajar con archivos.



---


# Agregar contenido con with y modo 'a'

```python
print("Agregando nueva l√≠nea con with")
with open("productos.txt", "a") as archivo:
    archivo.write("Cartuchera\n")
```

1. Usamos nuevamente with, esta vez en modo 'a' para agregar contenido al final del archivo sin borrar lo anterior.

2. Si el archivo no existe, se crea autom√°ticamente.

3. Cada llamada a write() agrega al final. No olvides el \n para que cada dato est√© en una l√≠nea distinta.

4. Al usar with, no es necesario llamar a close().


---



# Importante

1. El alias que se utiliza luego de as (por ejemplo, archivo) puede tener cualquier nombre. Su uso est√° limitado al bloque with; fuera de √©l, no ser√° reconocido.

2. El modo 'w' puede utilizarse dentro del bloque with igual que en una apertura convencional. Si el archivo no existe, lo crea. Si ya existe, sobrescribe completamente su contenido.
